```python
# Read the CSV file into a DataFrame

import pandas as pd

df = pd.read_csv(csv_file_path)
print(df)
```

```python
# Read the CVS file into a list of dictionaries

import csv

# List to store dictionaries
country_list = []

# Open the CSV file
with open(csv_file_path, newline='') as csvfile:
    # Create a CSV reader object with DictReader
    csvreader = csv.DictReader(csvfile)

    # Iterate through each row
    for row in csvreader:
      # Store each row as a dictionary in the list
      country_list.append(row)

# add Country key to the dictionaries

for country in country_list:
  country["Country"] = country.pop("")
  print(country)
```

```python
# Calculate the total number of countries

country_total = len(country_list)

print(country_total)
```

```python
# Calculate the total number of years

year_total = len(country_list[0]) - 1

print(year_total)
```

```python
# List the countries

for country in country_list:
  for key in country.keys():
    if key == "Country":
      print(country[key])
```

```python
# List the years

for key in country_list[0].keys():
  if key != "Country":
    print(key)
```

```python
# Calculate the total number of decisions

decision_total = 0

for country in country_list:
  for key in country.keys():
    if key != "Country":
      decision_total += int(country[key].replace(",", ""))

print(decision_total)
```

```python
# Function that calculates total number of decisions of a country

def country_decisions(country_name):
  country_decisions = 0
  for country in country_list:
    if country["Country"] == country_name:
      for key in country.keys():
        if key != "Country":
          country_decisions += int(country[key].replace(",", ""))
  return country_decisions

print(country_decisions("Zimbabwe"))
```

```python
# Funtion that calculates total number of decisions in a year
def year_decisions(year):
  year_decisions = 0
  for country in country_list:
    for key in country.keys():
      if key == year:
        year_decisions += int(country[key].replace(",", ""))
  return year_decisions

print(year_decisions("2019"))

```

```python
# Calculate the total number of decisions per country

# Dictionary to store values per country
country_decisions = {}

# Add countries and values to Dictionary
for country in country_list:
  country_decisions[country["Country"]] = 0
  for key in country.keys():
    if key != "Country":
      country_decisions[country["Country"]] += int(country[key].replace(",", ""))

# Print total number of decisions per country
for country in country_decisions:
  print(country, country_decisions[country])
```

```python
# Top 10 countries based on # of decisions

top10_countries = dict(sorted(country_decisions.items(), key=lambda item: item[1], reverse=True)[:10])

for country in top10_countries:
  print(country, top10_countries[country])
```

```python
# Graph top 10 countries

import matplotlib.pyplot as plt

# Sample data
countries = []
decisions = []
for key, value in top10_countries.items():
  countries.append(key)
  decisions.append(value)

# Plotting a bar chart
plt.bar(countries, decisions)

# Adding labels and title
plt.xlabel('Countries')
plt.ylabel('Decisions')
plt.title('Bar Plot')
plt.xticks(rotation=45, ha='right')

# Display the plot
plt.show()
```
